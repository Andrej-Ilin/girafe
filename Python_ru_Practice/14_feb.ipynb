```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + (y - np.cbrt(x**2))**2 - 1

#   
x = np.linspace(-1.5, 1.5, 300)
y = np.linspace(-1.5, 2, 300)
X, Y = np.meshgrid(x, y)

#   
Z = f(X, Y)

#  
plt.figure(figsize=(10, 10))
plt.contour(X, Y, Z, [0])  #     0
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('  $x^2+(y-\\sqrt[3]{x^2})^2=1$')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')  #     

plt.show()
```
