### 1 problem (4 points) 

Find and classify all the critical points of the following function

$$ f(x, y) = 7x - 8y + 2xy - x^2 + y^3 $$

### Solve

we will follow these steps:

- Calculate the first partial derivatives of ( f ) with respect to ( x ) and ( y ).
- Set the first partial derivatives equal to zero to find critical points.
- Classify the critical points using the second derivative test.


#### The first partial derivative with respect to ( x ) is:

$ f_x(x, y) = \frac{\partial f}{\partial x} = 7 + 2y - 2x. $

#### The first partial derivative with respect to ( y ) is:

$ f_y(x, y) = \frac{\partial f}{\partial y} = -8 + 2x + 3y^2. $

Step 2: Set the First Partial Derivatives to Zero

Setting the partial derivatives equal to zero gives us a system of equations:

( 7 + 2y - 2x = 0 )

( -8 + 2x + 3y^2 = 0 )

From the first equation, we can solve for ( y ):

$ 2y = 2x - 7 \implies y = x - \frac{7}{2}. $

Now we can substitute this expression for ( y ) into the second equation:

$ -8 + 2x + 3\left(x - \frac{7}{2}\right)^2 = 0. $

Expanding $ (x - \frac{7}{2})^2 $:

$ (x - \frac{7}{2})^2 = x^2 - 7x + \frac{49}{4}. $

Substituting back into the second equation:

$ -8 + 2x + 3\left(x^2 - 7x + \frac{49}{4}\right) = 0, $ 

$ -8 + 2x + 3x^2 - 21x + \frac{147}{4} = 0, $ 

$ 3x^2 - 19x + \left(-8 + \frac{147}{4}\right) = 0, $ 

$ 3x^2 - 19x + \frac{-32 + 147}{4} = 0, $ 

$ 3x^2 - 19x + \frac{115}{4} = 0.$

To eliminate the fraction, multiply the entire equation by 4:

$ 12x^2 - 76x + 115 = 0. $

Next, we can use the quadratic formula to solve for ( x ):

$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} = \frac{76 \pm \sqrt{(-76)^2 - 4 \cdot 12 \cdot 115}}{2 \cdot 12}$.

Calculating the discriminant:

$ b^2 - 4ac = 5776 - 5520 = 256. $

Now substituting back into the quadratic formula:

$ x = \frac{76 \pm 16}{24}. $

Calculating the two values of ( x ):

$( x_1 = \frac{92}{24} = \frac{23}{6} )$.

$( x_2 = \frac{60}{24} = \frac{5}{2} )$.

Finding Corresponding ( y ) Values

Substituting $( x_1 )$ and $( x_2 )$ back into $( y = x - \frac{7}{2} )$:

For $( x_1 = \frac{23}{6} ):$

 $ y_1 = \frac{23}{6} - \frac{21}{6} = \frac{2}{6} = \frac{1}{3}. $

For $( x_2 = \frac{5}{2} ):$

 $ y_2 = \frac{5}{2} - \frac{7}{2} = -1. $

Thus, we have two critical points:

$ \left( \frac{23}{6}, \frac{1}{3} \right) .$

$ \left( \frac{5}{2}, -1 \right) .$

Step 3: Classify the Critical Points

To classify these critical points, we need to compute the second partial derivatives:
$
( f_ = -2 ).

( f_ = 6y ).

( f_ = f_ = 2 ).
$

Now we apply the second derivative test. We form the Hessian determinant ( H ):

$ H = f_ - (f_)^2 $.

For Critical Point $ \left( \frac{23}{6}, \frac{1}{3} \right) :$

Calculating $( f_ )$:

$ f_ = 6 \left( \frac{1}{3} \right) = 2 $.

Now, calculate $( H )$:

$ H = (-2)(2) - (2)^2 = -4 - 4 = -8. $

Since $( H < 0 )$, this point is a saddle point.

For Critical Point $ \left( \frac{5}{2}, -1 \right) :$

Calculating $( f_ ):$

$ f_ = 6(-1) = -6. $

Now, calculate $( H ):$

$ H = (-2)(-6) - (2)^2 = 12 - 4 = 8. $

Since $( H > 0 ) and ( f_ < 0 )$, this point is a local maximum.

Summary of Critical Points

$ \left( \frac{23}{6}, \frac{1}{3} \right) $: Saddle point.

$ \left( \frac{5}{2}, -1 \right) $: Local maximum.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#  
def f(x, y):
    return 7 * x - 8 * y + 2 * x * y - x**2 + y**3

#     x  y
x = np.linspace(-2, 5, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

#   
critical_points = [(23/6, 1/3), (5/2, -1)]
critical_values = [f(cp[0], cp[1]) for cp in critical_points]

#  
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

#   
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

#   
for (cp_x, cp_y), cp_z in zip(critical_points, critical_values):
    ax.scatter(cp_x, cp_y, cp_z, color='red', s=100, label=f'Critical Point ({cp_x:.2f}, {cp_y:.2f})')

#  
ax.set_title('Surface Plot of f(x, y)')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('f(x, y)')
ax.legend()
plt.show()
```

### Problem 2. (6 points) 
Fitting a machine learning model means finding the optimal values of its parameters, which
comes down to optimizing some loss function L. Suppose you want to fit a linear regression model of
the form
$$y = w_0 + w_1\cdot x_1 + ... + w_n\cdot x_n$$

You want to do so by minimizing the following loss function:

$$\mathcal{L}(w) = \sum_{i=1}^{n} \left( y_i - \hat{y}_i\right)^2 + \lambda\cdot||w||^2$$

Here, $ \hat{y}_i$ is model's prediction for example $i$ and $||w||^2$ is the $l_2$ norm of the unknown weights vector $w = (w_0, w_1, ..., w_n)^T $.

The effect of adding this extra term to the loss function is that it forces us to chose small values for the unknown coefficients. The larger the value of the hyperparameter $\lambda$, the larger is the effect of regularization.


### (a) Find the gradient of $\mathcal{L}$. Hint: it might be useful to re-write $\mathcal{L}$ using matrix notation and use matrix calculus.

### Solution A:

$$\mathcal{L}(w) = \sum_{i=1}^{n} \left( y_i - \hat{y}_i\right)^2 + \lambda\cdot||w||^2$$

where:

- $ y_i $ is the true value,
- $ (\hat{y}i = w_0 + w_1 x{1i} + w_2 x_{2i} + \ldots + w_n x_) $ is the predicted value,
- $ w = (w_0, w_1, \ldots, w_n)^T $ is the vector of weights,
- $ \lambda $ is the regularization parameter.



We can express the predictions in matrix form. 

Let:

- $( \mathbf{X} )$ be the design matrix, which includes a column of ones for $( w_0 )$ and the input features $( x_1, x_2, \ldots, x_n )$.

- $( \mathbf{y} )$ be the vector of true outputs.
- $( \mathbf{w} )$ be the vector of weights.
Then we have:
$ \hat{\mathbf{y}} = \mathbf{X}\mathbf{w}. $

The loss function can then be expressed as:

$ L(\mathbf{w}) = |\mathbf{y} - \mathbf{X}\mathbf{w}|^2 + \lambda |\mathbf{w}|^2. $


#### Compute the Gradient

Using the properties of matrix calculus, we can compute the gradient of ( L ):

The term $( |\mathbf{y} - \mathbf{X}\mathbf{w}|^2 )$ can be expanded as:

$ |\mathbf{y} - \mathbf{X}\mathbf{w}|^2 = (\mathbf{y} - \mathbf{X}\mathbf{w})^T(\mathbf{y} - \mathbf{X}\mathbf{w}). $

Differentiating this with respect to $( \mathbf{w} )$:

$ \nabla_{\mathbf{w}} |\mathbf{y} - \mathbf{X}\mathbf{w}|^2 = -2\mathbf{X}^T(\mathbf{y} - \mathbf{X}\mathbf{w}). $

The term $( \lambda |\mathbf{w}|^2 )$ is straightforward:

$ \nabla_{\mathbf{w}} (\lambda |\mathbf{w}|^2) = 2\lambda \mathbf{w}. $

Putting these together, the gradient of $( L(\mathbf{w}) )$ is:

$ \nabla_{\mathbf{w}} L(\mathbf{w}) = -2\mathbf{X}^T(\mathbf{y} - \mathbf{X}\mathbf{w}) + 2\lambda \mathbf{w}. $

To simplify, we can factor out the 2:

$ \nabla_{\mathbf{w}} L(\mathbf{w}) = -\mathbf{X}^T(\mathbf{y} - \mathbf{X}\mathbf{w}) + \lambda \mathbf{w}. $

This gives us the gradient of the loss function with respect to the weights $( \mathbf{w} )$.

### (b) Using the gradient obtained above, derive the expression for the optimal model weights w that minimize the loss L.

To find the optimal weights $( \mathbf{w} )$ that minimize $ L(\mathbf{w}) $, we set the gradient to zero:

$ -\mathbf{X}^T(\mathbf{y} - \mathbf{X}\mathbf{w}) + \lambda \mathbf{w} = 0. $

Rearranging this equation gives:

$ \mathbf{X}^T(\mathbf{y} - \mathbf{X}\mathbf{w}) = \lambda \mathbf{w}. $

Expanding the left side:

$ \mathbf{X}^T\mathbf{y} - \mathbf{X}^T\mathbf{X}\mathbf{w} = \lambda \mathbf{w}. $

Combining the terms involving $( \mathbf{w} )$:

$ \mathbf{X}^T\mathbf{y} = \mathbf{X}^T\mathbf{X}\mathbf{w} + \lambda \mathbf{w}. $

Factoring out $( \mathbf{w} )$:

$ \mathbf{X}^T\mathbf{y} = (\mathbf{X}^T\mathbf{X} + \lambda I)\mathbf{w}, $

where $( I )$ is the identity matrix.

Now, we can solve for $( \mathbf{w} )$:

$ \mathbf{w} = (\mathbf{X}^T\mathbf{X} + \lambda I)^{-1}\mathbf{X}^T\mathbf{y}. $
