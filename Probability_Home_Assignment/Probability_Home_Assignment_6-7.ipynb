### Problem 1. (1 points)
 You have n enumerated letters and n enumerated envelopes. You
randomly put letters into envelopes. What is the expected value of the number of coinciding
numbers of the letter and its envelope?

#### Solution
$ n $: the number of letters/envelopes.

$ X $: the random variable representing the number of coinciding pairs (where a letter is placed in the envelope with the same index).


We define indicator random variables for each letter/envelope pair:

$$
X_i = 
\begin{cases} 
1 & \text{if letter } i \text{ is placed in envelope } i \\ 
0 & \text{otherwise} 
\end{cases}
$$

where $ i $ ranges from $ 1 $ to $ n $.

The total number of coinciding pairs $$ X $$ can be expressed as the sum of these indicator variables:

$$
X = X_1 + X_2 + \ldots + X_n
$$

***Step 1: Calculate the Expectation of Each Indicator Variable***

To find the expected value of $ X $, we first find the expected value of each indicator variable $ X_i $:

$$
E[X_i] = P(X_i = 1)
$$

For $ X_i = 1 $ to occur, letter $$ i $$ must be placed in envelope $ i $. Since the letters are placed randomly into the envelopes, the probability that any specific letter ends up in its corresponding envelope is:

$$
P(X_i = 1) = \frac{1}{n}
$$

***Step 2: Apply Linearity of Expectation***

Using the linearity of expectation, we can express the expected value of $ X $:

$$
E[X] = E[X_1 + X_2 + \ldots + X_n] = E[X_1] + E[X_2] + \ldots + E[X_n]
$$

Substituting the expected values we calculated:

$$
E[X] = E[X_1] + E[X_2] + \ldots + E[X_n] = n \cdot \frac{1}{n} = 1
$$

***Conclusion***

Thus, the expected value of the number of coinciding numbers of the letter and its envelope is:
$
\boxed{1}
$

This result indicates that if you randomly distribute $ n $ letters into $ n $ envelopes, on average, you can expect exactly one letter to end up in the correct envelope.

### Problem 2. (2 points)

Find the distribution of 
$
Z = X + Y
$
where $ X \sim \text{Pois}(\lambda_1) $ and $ Y \sim \text{Pois}(\lambda_2) $, and $ X \perp Y $.

***Solution***

Since $ X $ and $ Y $ are independent Poisson random variables, the sum $ Z = X + Y $ is also a Poisson random variable with parameter 
$
\lambda = \lambda_1 + \lambda_2.
$

Thus, we have:
$$
\boxed{Z \sim \text{Pois}(\lambda_1 + \lambda_2)}
$$


#### Problem 3. (2 points) Conditional Distribution of Poisson Variables


Find the distribution $ X | (X + Y) $ and its PMF $ P(X = k | X + Y = n) $.

***Solution***

Given that 
$
Z = X + Y \sim \text{Pois}(\lambda_1 + \lambda_2),
$

we can use the property of the conditional distribution of Poisson variables:

$$
P(X = k | Z = n) = \frac{P(X = k, Y = n - k)}{P(Z = n)}.
$$

Using the PMF of Poisson:
$$
P(X = k) = \frac{\lambda_1^k e^{-\lambda_1}}{k!}, \quad P(Y = n-k) = \frac{\lambda_2^{n-k} e^{-\lambda_2}}{(n-k)!},
$$

and 

$$
P(Z = n) = \frac{(\lambda_1 + \lambda_2)^n e^{-(\lambda_1 + \lambda_2)}}{n!}.
$$

Therefore, we have:

$$
P(X = k | Z = n) = \frac{\frac{\lambda_1^k e^{-\lambda_1}}{k!} \cdot \frac{\lambda_2^{n-k} e^{-\lambda_2}}{(n-k)!}}{\frac{(\lambda_1 + \lambda_2)^n e^{-(\lambda_1 + \lambda_2)}}{n!}}.
$$

Simplifying gives:

$$
P(X = k | Z = n) = \frac{n!}{k!(n-k)!} \cdot \left( \frac{\lambda_1}{\lambda_1 + \lambda_2} \right)^k \cdot \left( \frac{\lambda_2}{\lambda_1 + \lambda_2} \right)^{n-k}.
$$

This shows that 

$$
X | (X + Y = n) \sim \text{Binomial}(n, p)
$$
where 
$$
p = \frac{\lambda_1}{\lambda_1 + \lambda_2}.
$$


#### Problem 4

Find the expectation and variance of the exponential distribution $ \text{Exp}(\lambda) $.

***Solution***

The expected value and variance of an exponential distribution are given by:

***Expectation:***

$$
E[X] = \frac{1}{\lambda}
$$

***Variance:***

$$
\text{Var}(X) = \frac{1}{\lambda^2}
$$

For the case $ \lambda = 1 $:

$$
E[X] = 1 \quad \text{and} \quad \text{Var}(X) = 1
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
lambda_param = 1

# Define the range for x
x = np.linspace(0, 10, 1000)

# PDF of the exponential distribution
pdf = lambda_param * np.exp(-lambda_param * x)

# CDF of the exponential distribution
cdf = 1 - np.exp(-lambda_param * x)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF: $f(x) = \\lambda e^{-\\lambda x}$', color='blue')
plt.title('PDF of Exp(1)')
plt.xlabel('x')
plt.ylabel('Density')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF: $F(x) = 1 - e^{-\\lambda x}$', color='orange')
plt.title('CDF of Exp(1)')
plt.xlabel('x')
plt.ylabel('Probability')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

#### Problem 5. (2 points) Distribution of Minimum and Sum of Independent Exponential Variables

Consider independent random variables $ X \sim \text{Exp}(\lambda_X) $ and $ Y \sim \text{Exp}(\lambda_Y) $. 

1. Find the CDF of 
$
Z = \min(X, Y).
$

2. Let $ \lambda_X = \lambda_Y $. Find the CDF of 
$
Z = X + Y.
$

***Solution***

1. To find the CDF of $ Z = \min(X, Y) $, we have:
$$
F_Z(z) = P(Z \leq z) = P(\min(X, Y) \leq z) = 1 - P(X > z) P(Y > z).
$$

For exponential random variables, we know:
$$
P(X > z) = e^{-\lambda_X z}, \quad P(Y > z) = e^{-\lambda_Y z}.
$$

Thus, we can write:
$$
F_Z(z) = 1 - e^{-\lambda_X z} e^{-\lambda_Y z} = 1 - e^{-(\lambda_X + \lambda_Y) z}.
$$

This implies that:
$$
Z \sim \text{Exp}(\lambda_X + \lambda_Y).
$$

2. Now, let $ \lambda_X = \lambda_Y $. The sum of two independent exponential random variables follows a Gamma distribution:
$$
Z = X + Y \sim \text{Gamma}(k=2, \theta=\frac{1}{\lambda}).
$$

The CDF of $ Z $ can be derived using the convolution formula:
$$
F_Z(z) = P(X + Y \leq z).
$$

This is derived from the PDF of the Gamma distribution:
$$
f_Z(z) = \frac{\lambda^2 z e^{-\lambda z}}{(2 - 1)!} = \lambda^2 z e^{-\lambda z}, \quad (z \geq 0).
$$

Hence, the solution provides the CDF and characteristics of the distributions derived from the independent exponential random variables.

Let $ p $ be the probability of success and $ q = 1 - p $. The expected value is given by:
$$

$$
E[X] = \frac{1}{p}.
$$

***Variance of the Geometric Distribution***

Using the property of the variance:
$$
\text{Var}(X) = E[X^2] - (E[X])^2.
$$

Using the Law of the Unconscious Statistician (LOTUS), we can obtain:
$$
E[X^2] = \sum_{k=1}^{\infty} k^2 p q^{k-1}.
$$

After simplifications, we find:
$
E[X^2] = \frac{2 - p}{p^2},
$
leading to:
$$
\text{Var}(X) = \frac{q}{p^2},
$$
where $$ q = 1 - p $$.
