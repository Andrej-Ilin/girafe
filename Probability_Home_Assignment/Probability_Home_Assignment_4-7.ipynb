### Problem 1 (2 points)

Question: If $(X \sim Bi(n, p))$ and $(Y \sim Bi(m, p))$,

 and $(X \perp Y) (independent)$, what is $(P(X | X + Y = r))$, 
the PMF of the conditional distribution of $(X)$ given $(X + Y = r)$? What is the name of this distribution?


Solution:

Given (X) and (Y) are independent binomial random variables, their sum (X + Y) follows a binomial distribution:

$ X + Y \sim Bi(n+m, p) $

Using the property of conditional probability:

$
P(X = k | X + Y = r) = \frac{P(X = k \text{ and } Y = r-k)}{P(X + Y = r)}
$

The joint probability can be expressed as:


$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} $
$ P(Y = r-k) = \binom{m}{r-k} p^{r-k} (1-p)^{m-(r-k)} $


Thus,

$
P(X = k | X + Y = r) = \frac{\binom{n}{k} p^k (1-p)^{n-k} \cdot \binom{m}{r-k} p^{r-k} (1-p)^{m-(r-k)}}{P(X + Y = r)}
$

This leads to:

$
P(X = k | X + Y = r) = \frac{\binom{n}{k} \binom{m}{r-k}}{\binom{n+m}{r}}
$

This is the PMF of a Hypergeometric distribution with parameters (N = n + m), (K = n), and (n = r).

### Problem 2 (2 points)

Question: Find the PMF of the random variable (X), which is equal to the number of failures in a series of Bernoulli trials with success probability (p), carried out until there are (r) successes. This distribution is called Negative Binomial distribution.
Solution:

The PMF of the Negative Binomial distribution, which counts the number of failures before the (r)-th success, is given by:

$
P(X = k) = \binom{k+r-1}{r-1} p^r (1-p)^k, \quad k = 0, 1, 2, \ldots
$

where (p) is the probability of success and (1-p) is the probability of failure.

### Problem 5* (1 bonus point)

Question: 

Use Python package scipy.stats to plot the PMF of distribution from Problem 2 for different values of parameter (p).

Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

r = 5  # number of successes
p_values = $0.1, 0.5, 0.9$  # different values of p

x = np.arange(0, 30)  # Number of failures

for p in p_values:
    pmf = nbinom.pmf(x, r, p)
    plt.plot(x, pmf, label=f'p={p}')

plt.title('PMF of Negative Binomial Distribution')
plt.xlabel('Number of Failures')
plt.ylabel('Probability Mass Function')
plt.legend()
plt.show()
```

### Problem 6 (2 bonus points)

Question: 

A good student solves a problem correctly with probability 0.95, while a bad student — with probability 0.15. What is the minimal number of problems that the test should include so that the probability that a good student does not pass the test does not exceed 0.01, and the probability that the bad student passes the test does not exceed 0.1?

Approach:

Let (n) be the number of problems.

The good student passes if they solve more than (n/2). Thus: $ P(\text{good student fails}) \leq 0.01 $

The bad student passes if they solve more than (n/2): $ P(\text{bad student passes}) \leq 0.1 $

Using the Binomial distribution properties and solving these inequalities leads to finding the smallest (n).

```python
from scipy.stats import binom

# Define parameters
p_good = 0.95
p_bad = 0.15

n = 1
while True:
    # Good student fails
    prob_good_fails = binom.cdf(n//2, n, p_good)
    # Bad student passes
    prob_bad_passes = 1 - binom.cdf(n//2, n, p_bad)
    
    if prob_good_fails <= 0.01 and prob_bad_passes <= 0.1:
        break
    
    n += 1

print(f'Minimal number of problems: {n}')
```
