Check convexity of the following functions

### Task 1
#### Restating the Problem

We want to check the convexity of the function:

$$
h(x) = \frac{f(x)^2}{g(x)},
$$

where:
1. $f: \mathbb{R}^n \to \mathbb{R}$ is a nonnegative convex function ($f(x) \geq 0$ and convex).
2. $g: \mathbb{R}^n \to \mathbb{R}$ is a positive concave function ($g(x) > 0$ and concave).
3. Both $f$ and $g$ are twice continuously differentiable.

#### Convexity Analysis

##### Step 1: Restriction to a Line
Without loss of generality, we can restrict $f$ and $g$ to any line in $\mathbb{R}^n$. Convexity in higher dimensions can be verified by checking convexity of the restriction of $h(x)$ to all lines in the domain (since the epigraph of $h$ must remain convex under such restrictions). 

Thus, for $x, y \in \mathbb{R}$ (i.e., along a 1-dimensional slice), let:
$$
z = \theta x + (1-\theta)y, \quad \text{for } \theta \in [0, 1].
$$

##### Step 2: Convexity of $f^2$
Since $f(x)$ is nonnegative and convex, the square $f(x)^2$ is also convex. This follows because the square function $t \mapsto t^2$ is convex and monotonically increasing on $[0, \infty)$, and the composition of a convex function with a non-decreasing convex function preserves convexity. 

Thus:
$$
f(z)^2 \leq \left(\theta f(x) + (1-\theta)f(y)\right)^2.
$$

##### Step 3: Concavity of $g$
By the concavity of $g(x)$, we know:
$$
g(z) \geq \theta g(x) + (1-\theta)g(y).
$$

##### Step 4: Combining $f^2$ and $g$
We now analyze $h(z)$:
$$
h(z) = \frac{f(z)^2}{g(z)}.
$$

Using the inequalities derived for $f(z)^2$ and $g(z)$, we can bound $h(z)$:
$$
h(z) = \frac{f(z)^2}{g(z)} \leq \frac{\left(\theta f(x) + (1-\theta)f(y)\right)^2}{\theta g(x) + (1-\theta)g(y)}.
$$

##### Step 5: Convexity of the Function $\frac{u^2}{v}$
To conclude the proof, we need to verify the convexity of the map:
$$
(u, v) \mapsto \frac{u^2}{v}, \quad v > 0.
$$

This function is jointly convex in $(u, v)$ on the domain $\{(u, v) : v > 0\}$. Intuitively, this follows from the fact that:
- For fixed $v$, the function $u \mapsto u^2/v$ is convex (quadratic in $u$).
- For fixed $u$, the function $v \mapsto u^2/v$ is convex (since $-1/v$ is convex for $v > 0$).

Thus, by the joint convexity of $\frac{u^2}{v}$, we conclude:
$$
\frac{\left(\theta f(x) + (1-\theta)f(y)\right)^2}{\theta g(x) + (1-\theta)g(y)} \leq \theta \frac{f(x)^2}{g(x)} + (1-\theta) \frac{f(y)^2}{g(y)}.
$$

##### Step 6: Final Conclusion
Combining all the inequalities, we have:
$$
h(z) = \frac{f(z)^2}{g(z)} \leq \theta \frac{f(x)^2}{g(x)} + (1-\theta) \frac{f(y)^2}{g(y)}.
$$

This proves that $h(x) = \frac{f(x)^2}{g(x)}$ is convex.



#### Key Justifications
1. Convexity of $f^2$: Follows from the composition of a convex function ($f(x)$) with the convex, monotonic $t^2$ on $[0, \infty)$.
2. Joint convexity of $\frac{u^2}{v}$: A standard result in convex analysis, applicable for $v > 0$.

This completes the proof.

## Problem 2 (10 pts)

We consider a portfolio selection problem with $n$ assets and $N$ periods.
At the beginning of each period, we re-invest our total wealth, redistributing it over the $n$ assets using a fixed, constant, allocation strategy $x \in \mathbb{R}^n$, where $x \geq 0, \sum_{i=1}^n x_i = 1$.

In other words, if $W(t − 1)$ is our wealth at the beginning of period $t$, then during period $t$ we invest $x_i W(t−1)$ in asset $i$.
We denote by $\lambda(t)$ the total return during period $t$, i.e., $\lambda(t) = W(t)/W(t − 1)$.
At the end of the $N$ periods our wealth has been
multiplied by the factor $\prod_{i=1}^N \lambda(t)$.

We use a discrete stochastic model to account for the uncertainty in the returns. We assume that during each period there are $m$ possible scenarios, with probabilities $\pi_j, j = 1, \ldots, m$.
In scenario $j$, the return for asset $i$ over one period is given by $p_{ij}$ .
Therefore, the return $\lambda(t)$ of our portfolio during period $t$ is a random variable, with $m$ possible values $p_1^\top x, \ldots, p_m^\top x$ and distribution $\pi_j = \mathbb{P}(\lambda(t) = p_j^\top x)$.

The target for the investment strategy is to maximize the log long term relative wealth

$$ LogLTRW = \lim_{N \to \infty} \frac1N \log\left( \frac{W(N)}{W(0)} \right)$$

Task 1 (2 pts)

Show that $LogLTRW = \sum_{j=1}^m
\pi_j \log(p_j^\top x)$ according to the considered discrete stochastic model.

Task 2 (8 pts)

State the convex optimization problem that will provide strategy for maximization log long term relative wealth and solve it with cvxpy.
In the below code, matrix $P$ stores returns for $n=5$ assets and $m=10$ possible outcomes in each period. Also consider all outcomes equiprobable, so $\pi_j = 1/m$.

Compare the log long term relative wealth for the obtained optimal strategy woth the uniform strategy. Comment result.

In addition you can plot the possible trajectories sampled randomly from the optimal strategy ```x_opt``` and uniform strategy ```x_unif```.
This simulation assumes that $W(0) = 1$.

#### Solution to problem 2
Using the definition of wealth evolution:
$W(N) = W(0)\prod_{t=1}^N \lambda(t)$

Taking the log:
$\log(W(N)/W(0)) = \log(\prod_{t=1}^N \lambda(t)) = \sum_{t=1}^N \log(\lambda(t))$

Therefore:
$LogLTRW = \lim_{N \to \infty} \frac{1}{N}\sum_{t=1}^N \log(\lambda(t))$

By the law of large numbers, this converges to the expected value of $\log(\lambda(t))$:
$LogLTRW = E[\log(\lambda(t))]$

Given our discrete stochastic model:
$LogLTRW = \sum_{j=1}^m \pi_j \log(p_j^\top x)$

````python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

P = np.array(np.mat(
    '3.5000    1.1100    1.1100    1.0400    1.0100;\
     0.5000    0.9700    0.9800    1.0500    1.0100;\
     0.5000    0.9900    0.9900    0.9900    1.0100;\
     0.5000    1.0500    1.0600    0.9900    1.0100;\
     0.5000    1.1600    0.9900    1.0700    1.0100;\
     0.5000    0.9900    0.9900    1.0600    1.0100;\
     0.5000    0.9200    1.0800    0.9900    1.0100;\
     0.5000    1.1300    1.1000    0.9900    1.0100;\
     0.5000    0.9300    0.9500    1.0400    1.0100;\
     3.5000    0.9900    0.9700    0.9800    1.0100'))

m = P.shape[0]
n = P.shape[1]
x_unif = np.ones(n)/n; # uniform resource allocation

#################################################################
# This space is for your code to solve the stated optimization problem
# Store your result in the variable x_opt
pi = np.ones(m)/m

# Define variables and constraints
x = cp.Variable(n)
constraints = [
    cp.sum(x) == 1,  # sum of weights = 1
    x >= 0  # non-negative weights
]

# Objective function: maximize sum(pi_j * log(p_j^T x))
objective = cp.sum(cp.multiply(pi, cp.log(P @ x)))

# Define and solve the problem
prob = cp.Problem(cp.Maximize(objective), constraints)
prob.solve()

# Get optimal solution
x_opt = x

# Calculate LogLTRW for both strategies
logLTRW_opt = np.sum(pi * np.log(P @ x_opt.value))
logLTRW_unif = np.sum(pi * np.log(P @ x_unif))

print("Optimal portfolio weights:", x_opt.value)
print("Uniform portfolio weights:", x_unif)
print(f"LogLTRW (optimal strategy): {logLTRW_opt:.4f}")
print(f"LogLTRW (uniform strategy): {logLTRW_unif:.4f}")
# Создаем 10 пар графиков (5x2)
fig, axes = plt.subplots(5, 2, figsize=(15, 25))
fig.suptitle('Comparison of Optimal and Uniform Strategies', fontsize=16)
#################################################################

np.random.seed(10);
N = 10;  # number of random trajectories
T = 200; # time horizon
w_opt = np.zeros((N,T+1))
w_unif = np.zeros((N,T+1))
for i in range(N):
    events = np.floor(np.random.rand(T)*m)
    events = events.astype(int)
    P_event = P[events,:]
    w_opt[i,:] = np.append(1, np.cumprod(P_event @ x_opt.value))
    w_unif[i,:] = np.append(1, np.cumprod(P_event @ x_unif))
	
    #   Determining the position for the current plot
    row = i // 2
    col = i % 2
   
    # Plotting 
    axes[row, col].plot(range(T+1), w_opt[i,:], color="green", linewidth=2.0, label='Optimal')
    axes[row, col].plot(range(T+1), w_unif[i,:], color="red", linewidth=2.0, 
                      linestyle='--', label='Uniform')
    axes[row, col].set_yscale("log")
    axes[row, col].set_xlabel('Time')
    axes[row, col].set_ylabel('Wealth')
    axes[row, col].set_title(f'Trajectory {i+1}')
    axes[row, col].grid(True)
    axes[row, col].legend()

plt.tight_layout()
plt.show()

# Вывод средних значених благосостояния для обеих стратегий
print(f"Mean final wealth (Optimal): {np.mean(w_opt[:,-1]):.2f}")
print(f"Mean final wealth (Uniform): {np.mean(w_unif[:,-1]):.2f}")

# plt.gca().set_yscale('log')
# plt.plot(range(T+1), np.transpose(w_opt), color="green", linewidth=2.0)
# plt.plot(np.arange(T+1), np.transpose(w_unif), color="red", linewidth=2.0, linestyle='--')
# plt.yscale("log")
# plt.xlabel('time')
# plt.ylabel('wealth')
````

## Problem 3 (5 pts)

An oncology patient is given a dose of radiation $d_t \in \mathbb{R}_+$ in time periods $t = 1, \ldots, T − 1$, with the goal of shrinking a tumor to some specified target size while
minimizing the damage to the patient’s health.
We can choose the doses $d_t$, subject to the limit $d_t \leq d_{\max}$, where $d_{\max}$ is a given maximum dose.
We let $S_t \in \mathbb{R}_+$ denote the tumor size in period $t$.
The tumor size evolves as
$$S_{t+1} = \alpha \exp(−\beta d_t)S_t,$$
where $t=1,...,T −1$ and $\alpha > 1$ is the per-period tumor growth rate with no radiation, and $\beta > 0$ is a known constant.
(Since $d_t \geq 0$ and $\beta > 0$, we see that the radiation applied in one period shrinks the tumor in the next period.)
The initial tumor size $S_1$ is given. The goal is to achieve $S_T \leq S_{tar}$, where $S_{tar}$ is a target final tumor size.
We let $H_t \in \mathbb{R}_+$ denote some measure of the damage to the patient’s health from the radiation treatments.
It evolves as
$$ H_{t+1} = \gamma \exp(\delta d_t)H_t, \quad t=1, \ldots,T−1,$$
where $\gamma \in (0, 1]$ is the per-period damage recovery rate with no radiation, and $\delta > 0$ is a known constant.
(Since $d_t \geq 0$ and $\delta > 0$, we see that the radiation applied in one period increases the damage in the next period.)
The initial damage $H_1$ is given.
The goal is to find a series of doses $d_1, \ldots, d_{T −1}$ that satisfies the constraints described above, and minimizes the maximum damage $H_{\max} = \max_{t=1,\ldots,T} H_t$.

Report the optimal objective value, i.e., the maximum damage.
Plot the dose $d_t$, damage $H_t$, and tumor size $S_t$ versus t, for an optimal dose plan.

#### Solution to Problem 3

---

### Part (a): 

Formulating the Problem as a Convex Optimization Problem

#### Key Objectives:
1. Minimize the maximum damage:
   $$
   H_{\max} = \max_{t=1, \ldots, T} H_t
   $$
2. Ensure the tumor shrinks to the target size:
   $$
   S_T \leq S_{tar}
   $$

#### Constraints:
1. Dose constraints: $ 0 \leq d_t \leq d_{\max} $, $ t = 1, \ldots, T-1 $.
2. Tumor size evolution: 
   $$
   S_{t+1} = \alpha \exp(-\beta d_t) S_t, \quad t = 1, \ldots, T-1
   $$
3. Health damage evolution:
   $$
   H_{t+1} = \gamma \exp(\delta d_t) H_t, \quad t = 1, \ldots, T-1
   $$

#### Variables:
- Decision variables: $ d_t $ (radiation dose per time period).
- Auxiliary variables:
  - $ H_t $ (health damage).
  - $ S_t $ (tumor size).
  - $ H_{\max} $: The maximum damage.

#### Challenges and Relaxations:

1. Non-convexity in $ S_{t+1} = \alpha \exp(-\beta d_t) S_t $ and $ H_{t+1} = \gamma \exp(\delta d_t) H_t $:
   - These are nonlinear due to the product $ \exp(-\beta d_t) S_t $ and $ \exp(\delta d_t) H_t $.

2. Log-transformation for convexity:
   Define:
   - $ \hat{S}_t = \log(S_t) $,
   - $ \hat{H}_t = \log(H_t) $,
   - $ \hat{d}_t = d_t $.

   Then:
   $$
   \log(S_{t+1}) = \log(\alpha) - \beta d_t + \log(S_t),
   \quad \Rightarrow \quad \hat{S}_{t+1} = \log(\alpha) - \beta \hat{d}_t + \hat{S}_t.
   $$

   Similarly:
   $$
   \log(H_{t+1}) = \log(\gamma) + \delta d_t + \log(H_t),
   \quad \Rightarrow \quad \hat{H}_{t+1} = \log(\gamma) + \delta \hat{d}_t + \hat{H}_t.
   $$

   These relationships are now linear, which makes them convex.

3. Maximizing damage constraint:
   To minimize $ H_{\max} $, introduce a variable $ \hat{H}_{\max} $ such that:
   $$
   \hat{H}_t \leq \hat{H}_{\max}, \quad t = 1, \ldots, T.
   $$

4. Final tumor size constraint:
   $$
   \hat{S}_T \leq \log(S_{tar}).
   $$

#### Optimization Problem:
The transformed convex optimization problem is:
$$
\begin{aligned}
\text{minimize} \quad & \hat{H}_{\max} \\
\text{subject to} \quad
& \hat{S}_{t+1} = \log(\alpha) - \beta \hat{d}_t + \hat{S}_t, \quad t = 1, \ldots, T-1, \\
& \hat{H}_{t+1} = \log(\gamma) + \delta \hat{d}_t + \hat{H}_t, \quad t = 1, \ldots, T-1, \\
& \hat{H}_t \leq \hat{H}_{\max}, \quad t = 1, \ldots, T, \\
& \hat{S}_T \leq \log(S_{tar}), \\
& 0 \leq \hat{d}_t \leq d_{\max}, \quad t = 1, \ldots, T-1.
\end{aligned}
$$

This is now a convex optimization problem that can be solved efficiently using tools like CVXPY.

---

#### Part (b): 

Solving the Problem

#### Parameters:
- $ T = 20 $,
- $ d_{\max} = 1.2 $,
- $ \alpha = 1.05 $,
- $ \beta = 0.6 $,
- $ \gamma = 0.9 $,
- $ \delta = 0.3 $,
- $ S_1 = 1 $,
- $ S_{tar} = 0.01 $,
- $ H_1 = 1 $.

#### Implementation in pseudocode:

R> d <- Variable(T_val-1, nonneg = TRUE)

R> S_l <- Variable(T_val)

R> H_l <- Variable(T_val)

R> obj <- max(H_l)

R> constr <- list()

R> for(t in seq_len(T_val-1)) {

    + constr <- c(constr, list(S_l[t+1] == S_l[t] - beta*d[t] + log(alpha),
    + H_l[t+1] == H_l[t] + delta*d[t] + log(gamma)))

    + }

R> constr <- c(constr, list(S_l[1] == log(S1),

    + S_l[T_val] <= log(S_tar), H_l[1] == log(H1), d <= d_max))

R> prob <- Problem(Minimize(obj), constr)

R> result <- solve(prob)

````python
# Your solution is here

import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt

# Problem parameters
T = 20
d_max = 1.2
alpha = 1.05
beta = 0.6
gamma = 0.9
delta = 0.3
S1 = 1
S_tar = 0.01
H1 = 1

# Log-transformed constants
log_alpha = np.log(alpha)
log_gamma = np.log(gamma)
log_S_tar = np.log(S_tar)

# Variables
d = cp.Variable(T-1)  # doses
log_S = cp.Variable(T)  # log of tumor size
log_H = cp.Variable(T)  # log of health damage
log_H_max = cp.Variable()  # maximum health damage (log)

# Constraints
constraints = []
constraints.append(log_S[0] == np.log(S1))  # Initial tumor size
constraints.append(log_H[0] == np.log(H1))  # Initial health damage

for t in range(T-1):
    # Tumor size evolution
    constraints.append(log_S[t+1] == log_alpha - beta * d[t] + log_S[t])
    # Health damage evolution
    constraints.append(log_H[t+1] == log_gamma + delta * d[t] + log_H[t])
    # Dose constraints
    constraints.append(d[t] >= 0)
    constraints.append(d[t] <= d_max)

# Max damage constraint
for t in range(T):
    constraints.append(log_H[t] <= log_H_max)

# Final tumor size constraint
constraints.append(log_S[T-1] <= log_S_tar)

# Objective: Minimize max health damage
objective = cp.Minimize(log_H_max)

# Solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

# Extract results
d_opt = d.value
log_S_opt = log_S.value
log_H_opt = log_H.value
H_max_opt = np.exp(log_H_max.value)

# Tumor and health trajectories without treatment
log_S_no_treatment = [np.log(S1)]
log_H_no_treatment = [np.log(H1)]
for t in range(1, T):
    log_S_no_treatment.append(log_S_no_treatment[-1] + log_alpha)
    log_H_no_treatment.append(log_H_no_treatment[-1] + log_gamma)

log_S_no_treatment = np.array(log_S_no_treatment)
log_H_no_treatment = np.array(log_H_no_treatment)

# Plot results
time = np.arange(1, T+1)

plt.figure(figsize=(12, 8))

# Plot dose
plt.subplot(3, 1, 1)
plt.plot(time[:-1], d_opt, label="Optimal dose", marker='o')
plt.xlabel("Time")
plt.ylabel("Dose")
plt.legend()

# Plot tumor size
plt.subplot(3, 1, 2)
plt.plot(time, np.exp(log_S_opt), label="Tumor size (optimal)", marker='o')
plt.plot(time, np.exp(log_S_no_treatment), label="Tumor size (no treatment)", linestyle="--")
plt.xlabel("Time")
plt.ylabel("Tumor size")
plt.legend()

# Plot health damage
plt.subplot(3, 1, 3)
plt.plot(time, np.exp(log_H_opt), label="Health damage (optimal)", marker='o')
plt.plot(time, np.exp(log_H_no_treatment), label="Health damage (no treatment)", linestyle="--")
plt.xlabel("Time")
plt.ylabel("Health damage")
plt.legend()

plt.tight_layout()
plt.show()

print(f"Optimal H_max: {H_max_opt:.4f}")
````
