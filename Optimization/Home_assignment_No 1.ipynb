```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Define the function
def f(x):
    return np.abs(x * np.cos(x))

def f_deriv(x):
    return np.cos(x) - x * np.sin(x) 

# Define the range for x
x = np.linspace(-4, 4, 400)
y = f(x)  
y_deriv = f_deriv(x)  # Derivative of f

# Find the roots of f_deriv
extremum_points = fsolve(f_deriv, [-3, -1, 1, 3])  # Initial guesses based on the expected locations of extrema
extremum_values = f(extremum_points)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=r'$f(x) = |x \cos x|$', color='blue')
plt.plot(x, y_deriv, label=r"$f'(x)$", color='orange', linestyle='--')

# Mark the extremum points on the plot
plt.scatter(extremum_points, extremum_values, color='red', zorder=5)
for point in zip(extremum_points, extremum_values):
    plt.annotate(f'({point[0]:.2f}, {point[1]:.2f})', 
                 xy=point, 
                 textcoords='offset points', 
                 xytext=(0,10), 
                 ha='center', 
                 color='red')

plt.title('Plot of the function $f(x) = |x \cos x|$ and its derivative')
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.xlim(-4, 4)
plt.ylim(-5, 5)
plt.show()
```
