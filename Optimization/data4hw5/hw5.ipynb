### Part 1: choose the location  for a new warehouse

- (4 pts) An urban planner would like to choose the location $x \in \mathbb{R}^2$ for a new warehouse.
This should be close to $n$ distribution centers located at $y_1, \ldots, y_n \in \mathbb{R}^2$. The objective function to identify the location of the new warehouse is the worst-case distance.
State the optimization problem that helps to identify the proper location.

### solution:

To help the urban planner find the optimal location $ x $ for the new warehouse, we are tasked with minimizing the worst-case (maximum) distance to the distribution centers. The goal is to identify $ x \in \mathbb{R}^2 $ that minimizes the largest Euclidean distance between $ x $ and any of the distribution centers $ y_1, y_2, \dots, y_n $.

***Mathematical Formulation***

Let the location of the warehouse be $ x = (x_1, x_2) \in \mathbb{R}^2 $, and the locations of the distribution centers be $ y_1, y_2, \dots, y_n \in \mathbb{R}^2 $. The Euclidean distance between the warehouse location $ x $ and a distribution center $ y_i $ is given by:
$$ d(x, y_i) = \sqrt{(x_1 - y_{i1})^2 + (x_2 - y_{i2})^2} $$
where $ y_i = (y_{i1}, y_{i2}) $ for $ i = 1, 2, \dots, n $.

The objective is to minimize the worst-case (maximum) distance, which means we want to minimize the maximum value of $ d(x, y_i) $ across all distribution centers $ i $. Therefore, the optimization problem can be formulated as:
$$ \min_{x \in \mathbb{R}^2} \max_{i = 1, 2, \dots, n} \left( \sqrt{(x_1 - y_{i1})^2 + (x_2 - y_{i2})^2} \right) $$

***Objective:***

The objective function is:
$$ \min_{x \in \mathbb{R}^2} \max_{i} d(x, y_i) $$

This problem is known as $\textit{min-max}$ optimization problem, where the urban planner seeks the location $ x $ that minimizes the maximum distance to any of the given distribution centers.

The start-up will lose money on the campaign if $x_{mix} < 0$, and will consider the campaign successful if $x_{mix} > L$. Alice and Bob’s goal is to maximize the probability of a successful campaign while ensuring the probability of a loss is no more than $\beta$.

#### Solution a: ***Since we don't have any specific data, we can only make an approximation.***


##### Definitions and Problem Setup

1. **Ad Options and Revenue Distribution**: 
   - There are $ n $ ads available.
   - Each ad $ j $ generates revenue $ x_j $ with a discrete probability distribution over $ m $ possible revenue outcomes $ c_1, c_2, \dots, c_m \in \mathbb{R} $, where $ c_1 < c_2 < \dots < c_m $.
   - The probability of the revenue outcome $ x_j = c_i $ for ad $ j $ is given by $ P_{ij} $. Thus, $ P \in \mathbb{R}^{m \times n} $ is the matrix describing the distributions for all ads.

2. **Randomized Strategy**: 
   - In Alice's randomized strategy, each ad $ j $ is shown with a probability $ \theta_j $, where $ \theta \in \mathbb{R}^n_+ $ and $ \sum_{j=1}^n \theta_j = 1 $.
   - The mixed revenue $ x_{\text{mix}} $ follows the distribution $ p = P\theta $, where $ p_i = P(x_{\text{mix}} = c_i) $ is the probability of revenue $ c_i $ occurring.

3. **Success and Failure Criteria**: 
   - The campaign is **successful** if the revenue per view $ x_{\text{mix}} > L $.
   - The campaign will be considered a **failure** if $ x_{\text{mix}} < 0 $.
   - The **probability of a loss** is $ P(x_{\text{mix}} < 0) $, and the goal is to ensure this probability is no more than $ \beta $.
   - The **goal** is to maximize the probability of success $ P(x_{\text{mix}} > L) $, subject to the constraint on the probability of loss.

---

#### Formulating the Optimization Problem

Let the distribution of the mixed revenue $ x_{\text{mix}} $ be $ p = P\theta $, where $ p_i = P(x_{\text{mix}} = c_i) $ represents the probability of $ x_{\text{mix}} = c_i $.

The optimization problem can be stated as follows:

$$
\max_{\theta \in \mathbb{R}^n_+} \, P(x_{\text{mix}} > L)
$$
subject to the constraint:
$$
P(x_{\text{mix}} < 0) \leq \beta
$$
and the normalization condition:
$$
\sum_{j=1}^n \theta_j = 1.
$$

#### Step-by-Step Breakdown

1. **Revenue Distribution**: 
   - The revenue per view $ x_{\text{mix}} $ is a random variable with a probability distribution given by the vector $ p = P\theta $, where each component $ p_i $ represents the probability that $ x_{\text{mix}} = c_i $.

2. **Probability of Success**:
   - The success condition is $ x_{\text{mix}} > L $, so the probability of success is the sum of the probabilities for $ c_i > L $:
   $$
   P(x_{\text{mix}} > L) = \sum_{i: c_i > L} p_i.
   $$

3. **Probability of Loss**:
   - The failure condition is $ x_{\text{mix}} < 0 $, so the probability of a loss is the sum of the probabilities for $ c_i < 0 $:
   $$
   P(x_{\text{mix}} < 0) = \sum_{i: c_i < 0} p_i.
   $$

4. **Objective**: 
   - Maximize the probability of success subject to the constraint on the probability of loss.

#### Final Optimization Problem

The problem can now be written as:

$$
\max_{\theta \in \mathbb{R}^n_+} \sum_{i: c_i > L} (P\theta)_i
$$
subject to:
$$
\sum_{i: c_i < 0} (P\theta)_i \leq \beta
$$
and the constraint that:
$$
\sum_{j=1}^n \theta_j = 1.
$$

---

### Interpretation

- The vector $ \theta = (\theta_1, \theta_2, \dots, \theta_n) $ represents the fraction of time each ad is displayed under Alice's randomized policy.
- The objective is to find the optimal distribution of $ \theta $ to maximize the probability of success (i.e., the probability that the revenue per view exceeds $ L $) while ensuring that the probability of a loss (i.e., the probability that the revenue per view is less than 0) is at most $ \beta $.
- The constraint $ \sum_{j=1}^n \theta_j = 1 $ ensures that the total probability is normalized, meaning that the ads are shown with a total probability of 1.

This formulation allows Alice and Bob to analyze the trade-off between success and risk in the randomized ad strategy.

#### Solution b: We know the data and that:
Using the data $n$, $m$, $P$, $c$, $\beta$, and $L$ in rand_policy_data.*, determine whether it is better to

````python
from scipy.io import savemat
import numpy as np

# Define the data
n = 4
m = 5
c = np.array([-2, -1, 0, 1, 2])
P = np.array([
  [.1, 0.4, 0, 0, .5],
  [.3, 0, 0, 0, .7],
  [.3, 0, 0, .4, .3],
  [0, 0, .75, .25, 0],
]).T
beta = 0.1
L = 0

# Save the data to a .mat file
data = {"n": n, "m": m, "c": c, "P": P, "beta": beta, "L": L}
savemat("rand_policy_data.mat", data)

print("rand_policy_data.mat file has been created.")
````

````python
#Step 1: Load the Data
import scipy.io as sio
import numpy as np
import cvxpy as cp

# Load data from rand_policy_data.mat
data = sio.loadmat("rand_policy_data.mat")

# Extract values
n = data["n"][0][0]  # Number of ads
m = data["m"][0][0]  # Number of revenue outcomes
P = data["P"]  # Probability matrix (m x n)
c = data["c"].flatten()  # Revenue outcomes (m,)
beta = data["beta"][0][0]  # Maximum allowed loss probability
L = data["L"][0][0]  # Success threshold
````

````python
# Single ad analysis
best_single_success = 0
best_single_ad = None

for j in range(n):
    # Probability of success for ad j
    success_prob = np.sum(P[c > L, j])
    # Probability of loss for ad j
    loss_prob = np.sum(P[c < 0, j])
    
    if loss_prob <= beta and success_prob > best_single_success:
        best_single_success = success_prob
        best_single_ad = j

print("Best Single Ad:", best_single_ad)
print("Best Single Ad Success Probability:", best_single_success)
````

````python
# Define variables for the randomized policy
theta = cp.Variable(n)

# Define the mixed distribution probabilities
p = P @ theta

# Success and loss probabilities
success_prob = cp.sum(p[c > L])
loss_prob = cp.sum(p[c < 0])

# Constraints
constraints = [
    loss_prob <= beta,
    cp.sum(theta) == 1,
    theta >= 0
]

# Objective: Maximize success probability
objective = cp.Maximize(success_prob)

# Solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

# Results
randomized_success = success_prob.value
optimal_theta = theta.value

print("Randomized Success Probability:", randomized_success)
print("Optimal Theta (Randomized Policy):", optimal_theta)
````

````python
if best_single_success > randomized_success:
    print("The Single Ad Policy is Better.")
else:
    print("The Randomized Policy is Better.")
````

### Part 3: Radiation Treatment Optimization Problem
- (5 pts) In radiation treatment, radiation is delivered to a patient, with the
goal of killing or damaging the cells in a tumor, while carrying out minimal damage to other tissue.
The radiation is delivered in beams, each of which has a known pattern; the level of each beam can
be adjusted.
(In most cases multiple beams are delivered at the same time, in one ‘shot’, with the
treatment organized as a sequence of ‘shots’.)
We let $b_j$ denote the level of beam $j$, for $j = 1, \ldots, n$.
These must satisfy $0 \leq b_j \leq B_{max}$, where $B_{max}$ is the maximum possible beam level.
The exposure
area is divided into $m$ voxels, labeled $i = 1,\ldots, m$.
The dose $d_i$ delivered to voxel $i$ is linear in
the beam levels, i.e., $d_i = \sum_{j=1}^n A_{ij} b_j$ .
Here $A \in \mathbb{R}^{m \times n}$ is a (known) matrix that characterizes the beam patterns.
We now describe a simple radiation treatment planning problem.
A (known) subset of the voxels, $T \subset \{1, \ldots, m\}$, corresponds to the tumor or target region.
We require that a minimum radiation dose $D_{target}$ be administered to each tumor voxel, i.e., $d_i \geq D_{target}$ for $i \in T$.
For all other voxels, we would like to have $d_i \leq D_{other}$, where $D_{other}$ is a desired maximum dose for non-target voxels. This is generally not feasible, therefore you should identify the beams levels such that the damage of the non-target voxels will be as small as possible.
State the corresponding optimization problem and explain the choice of the objective function.



In this case, we are optimizing the beam levels $ b_j $ to deliver radiation to a tumor while minimizing the damage to non-tumor regions.

**Definitions:**
- $ b_j $: the level of beam $ j $, $ 0 \leq b_j \leq B_{\text{max}} $.
- $ d_i = \sum_{j=1}^n A_{ij} b_j $: the dose delivered to voxel $ i $, where $ A $ is a matrix describing the beam patterns.
- $ T $: the set of tumor voxels.
- $ D_{\text{target}} $: minimum required dose for tumor voxels.
- $ D_{\text{other}} $: desired maximum dose for non-target voxels.

We want to satisfy the following conditions:
1. For each voxel $ i \in T $, the dose must satisfy $ d_i \geq D_{\text{target}} $.
2. For each voxel $ i \notin T $, the dose must satisfy $ d_i \leq D_{\text{other}} $, but minimizing this damage is the goal if it’s not feasible to meet the desired dose everywhere.

The objective function should minimize the total damage to non-target voxels, which can be modeled as the excess dose over the desired maximum $ D_{\text{other}} $.

Thus, the optimization problem is:

$$
\min_{b_1, \ldots, b_n} \sum_{i \notin T} \max(0, d_i - D_{\text{other}})
$$

subject to:
$$
d_i \geq D_{\text{target}} \quad \forall i \in T
$$
$$
0 \leq b_j \leq B_{\text{max}} \quad \forall j = 1, \ldots, n
$$

This problem minimizes the dose excess for non-target voxels while ensuring that tumor voxels receive the necessary dose.

````python
# Your solution is here

import cvxpy as cp
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the dataset
demand_points = pd.read_csv('SF_demand_205_centroid_uniform_weight.csv')
target_demand_points = demand_points[:30]  # Extract a subset of demand points

# Extract the coordinates of the demand points
y = target_demand_points[['long', 'lat']].values
n = y.shape[0]

# Define the variables
x = cp.Variable(2)  # The location of the warehouse (2D)
t = cp.Variable()   # The maximum distance to minimize

# Define the constraints
constraints = [cp.norm(x - y[i], 2) <= t for i in range(n)]

# Define the objective function
objective = cp.Minimize(t)

# Formulate and solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

# Optimal location
optimal_location = x.value
print("Optimal Warehouse Location:", optimal_location)

# Plotting the results
gdf = gpd.GeoDataFrame(
    target_demand_points, 
    geometry=gpd.points_from_xy(target_demand_points['long'], target_demand_points['lat'])
)
fig, ax = plt.subplots(figsize=(8, 8))
gdf.plot(ax=ax, color='blue', label='Demand Points')
ax.scatter(optimal_location[0], optimal_location[1], color='red', label='Optimal Warehouse Location')
ax.legend()
plt.show()
````

````python
# this one is randomly generated; a real one would use the beam geometry
import scipy.io as sio
import numpy as np
# Your solution is here

def get_value(mat, name):
    value = mat[name]
    return value[0][0] if np.shape(value) == (1, 1) else value

mat = sio.loadmat("treatment_planning_data.mat")
n = get_value(mat, "n")  # Number of beams
mtarget = get_value(mat, "mtarget")  # Number of tumor voxels
mother = get_value(mat, "mother")  # Number of other voxels
Atarget = get_value(mat, "Atarget")  # Tumor beam pattern matrix
Aother = get_value(mat, "Aother")  # Non-tumor beam pattern matrix
Bmax = get_value(mat, "Bmax")  # Maximum beam level
Dtarget = get_value(mat, "Dtarget")  # Required dose for tumor voxels
Dother = get_value(mat, "Dother")  # Desired max dose for non-tumor voxels

# Define variables
b = cp.Variable(n)  # Beam levels
e = cp.Variable(mother, nonneg=True)  # Excess dose for non-tumor voxels

# Constraints
constraints = []

# Tumor voxel constraints
for i in range(mtarget):
    constraints.append(Atarget[i, :] @ b >= Dtarget)

# Non-tumor voxel constraints (excess dose)
for i in range(mother):
    dose_other = Aother[i, :] @ b
    constraints.append(e[i] >= dose_other - Dother)  # Excess dose
    constraints.append(e[i] >= 0)  # Non-negativity of excess dose

# Beam level constraints
constraints.extend([0 <= b, b <= Bmax])

# Objective: Minimize total excess dose
objective = cp.Minimize(cp.sum(e))

# Solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

# Results
print("Total excess dose:", np.sum(e.value))
print()
print("Optimal beam levels (b):", b.value[:10])
````
